### Balancemento do Dataset

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=42)#random_state=42
```

#### Balanceamento do Long

```python
X_res, y_res = rus.fit_resample(treino_x, treino_y)
X_res_s, z_res_s = rus.fit_resample(treino_x, treino_z)
```

```python
z_res_s.value_counts()
```

```python
if(hiper == True):
    # Define the SVM model
    svm_model = SVC()

    # Define the parameter grid
    param_grid = {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': [0.1, 0.5, 1]
    }

    # Perform grid search with cross-validation
    grid_search = GridSearchCV(estimator=svm_model, param_grid=param_grid, verbose = 2)
    grid_search.fit(X_res, y_res)

    print('SVM: ', grid_search.best_params_)

    # Print the best parameters and best score
    print("Best Parameters: ", grid_search.best_params_)
    print("Best Score: ", grid_search.best_score_)

    # Evaluate the model on the test set
    accuracy = grid_search.score(teste_x, teste_y)
    print("Test Accuracy: ", accuracy)

elif(hiper == False):
    svm_model = SVC()
```
